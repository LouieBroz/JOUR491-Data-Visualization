# Aesthetics

Aesthetics on charts is important. And getting familiar with what you can do with them is important too. For an idea of what is important, [go here](http://stephanieevergreen.com/wp-content/uploads/2014/05/DataVizChecklist_May2014.pdf). 

We need to clean up our charts. We have loads of aesthetic options. Let's get back to our orinal chart of enrollment growth at UNL. 

```python
import pandas as pd
from ggplot import *
%matplotlib inline

df = pd.read_csv('../../Data/enrollment.csv')

ggplot(df, aes(x='Year', y='Enrollment')) + geom_line() 
```

First thing first, we need labels. There's several ways we can do this. The simplest is `labs` which is short for labels. In labs, you'll define the x axis label, the y axis label and the title. 

```python
ggplot(df, aes(x='Year', y='Enrollment')) + geom_line() + labs(x="Academic year", y="Fall enrollment", title="Enrollment at UNL since 1967")
```

Now, we should fix the scales. The defaults truncate the scale to the lowest and the highest. What effect does that have?

So to solve that, we need to create our own limits by overriding the defaults. With our data, we have two kinds of scale: continuous and date. 

We can scale the continuous using a continuous scale and passing in limits:

`scale_y_continuous(limits=(10000,28000))`

And we can do the same with dates like this: 

`scale_x_date(limits=(1967, 2016))`

```python
ggplot(df, aes(x='Year', y='Enrollment')) + geom_line() + labs(x="Academic year", y="Fall enrollment", title="Enrollment at UNL since 1967") + scale_y_continuous(limits=(10000,28000)) + scale_x_date(limits=(1967, 2016)) 
```

Looking a lot better. Now we can apply themes. There are some canned themes: `+ theme_bw()`

Or we can override things with our own themes. There are dozens of options [you can see here](https://github.com/yhat/ggplot/blob/master/ggplot/themes/theme.py). 

```python
ggplot(df, aes(x='Year', y='Enrollment')) + geom_line() + labs(x="Academic year", y="Fall enrollment", title="Enrollment at UNL since 1967") + scale_y_continuous(limits=(10000,28000)) + scale_x_date(limits=(1967, 2016)) + theme(x_axis_text=element_text(color="blue"))
```

```python
ggplot(df, aes(x='Year', y='Enrollment')) + geom_line() + labs(x="Academic year", y="Fall enrollment", title="Enrollment at UNL since 1967") + scale_y_continuous(limits=(10000,28000)) + scale_x_date(limits=(1967, 2016)) + theme(x_axis_text=element_text(color="blue")) + theme_bw()
```

```python
ggplot(df, aes(x='Year', y='Enrollment')) + geom_line(color="red") + labs(x="Academic year", y="Fall enrollment", title="Enrollment at UNL since 1967") + scale_y_continuous(limits=(10000,28000)) + scale_x_date(limits=(1967, 2016)) + theme(x_axis_text=element_text(color="black")) + theme_bw()
```
