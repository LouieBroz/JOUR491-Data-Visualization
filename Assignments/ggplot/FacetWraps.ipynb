```python
import pandas as pd
from ggplot import *
%matplotlib inline

df = pd.read_csv('../../Data/collegeenrollmentredux.csv')

majors = df[df["Count"] != 0]

majors = df[df["College"] != "Visitors"]
majors = majors[majors["College"] != "Explore Center Undergraduate"]
```

```python
majors.head()
```

```python
ggplot(majors, aes(x='Total', y="Count", color="RaceGender")) + geom_point() + xlab("Total Students") + ylab("Students in the Major") + ggtitle("Major by Race and Sex At UNL") + facet_wrap(y='College')
```
