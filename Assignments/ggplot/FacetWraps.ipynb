```python
import pandas as pd
from ggplot import *
%matplotlib inline

df = pd.read_csv('../../Data/collegeenrollmentredux.csv')

majors = df[df["Count"] != 0]
majors = majors[majors["College"] == "College of Arts and Sciences"]

#majors = df[df["College"] != "Visitors"]
#majors = majors[majors["College"] != "Explore Center Undergraduate"]
```

```python
majors.head()
```

```python
ggplot(majors, aes(x='MajorName', y="Count", color="RaceGender")) + geom_bar() + xlab("Major") + ylab("Students in the Major") + ggtitle("Major by Race and Sex At UNL") + facet_wrap(x='MajorName', y='College')
```
