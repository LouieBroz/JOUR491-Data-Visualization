# Starting with ggplot

ggplot2 is a wildly popular charting library in R. R is a general purpose statistics language that has its devotees and its detractors. I find it hard to learn as a beginner, unlike Python, and fortunately someone has ported much of ggplot2 to Python. So we're going to use it. You'll need to install it, so from within your homework environment, on the command line, you need to `pip install ggplot`

[Here is the documentation](http://ggplot.yhathq.com/docs/index.html). There's more, including more code, on GitHub here. 

The first thing is that ggplot is made to work with Pandas, a different data analysis library than Agate. It too has its devotees and detractors. I find it way more tool than most data journalists need, but it's easy enough to read data into a Pandas data frame and go from there. First, some setup. 

```python
import pandas as pd
from ggplot import *
%matplotlib inline
```

Now just import the data into a dataframe called df. It's very similar to Agate if you read it. To start, we're going to make a simple line graph using some enrollment data I have laying around for UNL. It's in the Data folder on Github.

```python
df = pd.read_csv('../../Data/enrollment.csv')
```

To see the headers in your data, which is handy, and the first five records, just do `df.head()`

```python
df.head()
```

Now, ggplot can be a little frustrating at first because it's not very explict what's going on. But it's very efficient, meaning not much code is needed to make something. 

So to make a line graph, you do this:

```python
ggplot(df, aes(x='Year', y='Enrollment')) + geom_line(color="red", size = 20)
```

Let's unpack that a bit. 

`ggplot(df, aes(x='Year', y='Enrollment')) + geom_line(color="red", size = 20)`

So we start with ggplot and pass it some information. The first thing is our data frame, called df. Then we pass it aes, which is short for aesthetics. In this case, we're not doing much: we're passing in the x and y variables, explicitly stated. You can do it implicitly too. Then we add the type of chart we're making -- geom_line. In geom_line, we can give it some instructions, like the color and the size. 

Bar charts are easy too. You don't even need to do group bys. 

```python
df2 = pd.read_csv('../../Data/mountainlions.csv')
df2.head()
```

```python
ggplot(aes(x='Cofirm Type'), data=df2) + \
     geom_bar()
```
